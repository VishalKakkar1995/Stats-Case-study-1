```python
import pandas as pd
import numpy as np
import math
import scipy.stats as stats
```

```python
Diet_Study=pd.read_csv("C:/Users/Vishal/Desktop/Python/7. Basic Stats - Case Study-1/dietstudy.csv")

Diet_Study.head(2)
```

```python
print("1. In order to check if there has been any change we'll check the data for tg0 and tg4. For weight we'll be checking the data for wgt0 and wgt4")

print("\nFor the null hypothesis we'll be considering that for both the intervals the values are same ")

print("\nFor weight using the values of wgt0 and wgt4\n")

print(stats.ttest_rel(a=Diet_Study.wgt0,b=Diet_Study.wgt4))

print("\nWe can see that the P value is very small so our null hypothesis of having the two values equal doesn't hold any true and also after seeing the mean of the two values")

print("\n Mean of WGT0=",Diet_Study.wgt0.mean())
print(" Mean of WGT4=",Diet_Study.wgt4.mean())

print('We can clearly conclude that the weight of the patients have reduced after following the new diet')




print("\nFor triglyceride level using the values of tg0 and tg4\n")

print(stats.ttest_rel(a=Diet_Study.tg0,b=Diet_Study.tg4))

print("\nWe can see that the P value is not so small so our null hypothesis of having the two values equal does hold true and we can say that the triglyceride lecel didn't change that much even with the changed diet and also after seeing the mean of the two values")

print("\n Mean of TG0=",Diet_Study.tg0.mean())
print(" Mean of TG4=",Diet_Study.tg4.mean())

print('We can clearly conclude that the triglyceride level of the patients didnot change substatitally even after following the new diet')
```

```python
Credit_promo=pd.read_csv("C:/Users/Vishal/Desktop/Python/7. Basic Stats - Case Study-1/creditpromo.csv")

Credit_promo.head(2)
```

```python
Standard=Credit_promo.dollars[(Credit_promo['insert']=='Standard')]

Promotion=Credit_promo.dollars[(Credit_promo['insert']=='New Promotion')]
```

```python
print("2.In order to check if the promotion changed the sales we'll be initially considering that both the types have equal average and this will be our null hypothesis")

print("\nSince the data is of two independent sections we'll be using the 2 sample test\n")

print(stats.ttest_ind(a=Standard,b=Promotion,equal_var=True))

print('\nSince most of the tests are done at 95% accuracy and for that the p value should be greater than 0.05 but in this case the vale of P is less than that so our assumption that both the averages are same turns out to be false and by seeing the mean of both dataframes we can conlude that promotion did increase the sales of the company')

print("\nMean of Standard customers=",Standard.mean())
print("Mean of Promotion customers=",Promotion.mean())
```

```python
Pollination=pd.read_csv("C:/Users/Vishal/Desktop/Python/7. Basic Stats - Case Study-1/pollination.csv")

Pollination.head()
```

```python
print("3.a In order to check if the overall seed yield is 200 we'll be considering the null hypothesis where we'll be testing for Ho=200")

print("\nAs we are given it to check for 200 we'll be using one samle t test\n")

print(stats.ttest_1samp(a=Pollination.Seed_Yield_Plant,popmean=200))

print("\nUsually we do the test for 95% accuract and for that the P value should be greater than 0.05. In this case since the P value is less than that we'll conclude that our null hypothesis of average to be 200 doesn't hold true. on checking the mean we can clearly see that the value is less than 200")

print("The average value to Seeds is =",Pollination.Seed_Yield_Plant.mean())
```

```python
print("3.b In order to check if the  natural pollination and hand pollination under open field conditions are equally effective or are significantly different we'll be doing the independent t test as both samples are exclusive from each other on each column to check the on all parameters")

Natural=Pollination[Pollination.Group=='Natural']
Hand=Pollination[Pollination.Group=='Hand']

print("\nIn order to check for the fruit yield we'll be performing the 2 sample test on fruit wt column for both data\n")

print(stats.ttest_ind(a=Natural.Fruit_Wt,b=Hand.Fruit_Wt,equal_var=True))

print("\nBased on the P value we can conclude that the hypothesis that we had earlier started with that both the values have same average is false as the P value is very small. On checking the mean of the two")

print("\nThe mean for Naturally grown fruits =",round(Natural.Fruit_Wt.mean(),2))
print("The mean for Hand pollination grown fruits =",round(Hand.Fruit_Wt.mean(),2))

print("\nOn seeing the mean we can clearly see that the hand pollinated plants give better fruits")


print("----------------------------------------------------------------------------------------------------------------------")


print("\nIn order to check for the fruit yield we'll be performing the 2 sample test on Seed Yield of plants column for both data\n")

print(stats.ttest_ind(a=Natural.Seed_Yield_Plant,b=Hand.Seed_Yield_Plant,equal_var=True))

print("\nBased on the P value we can conclude that the hypothesis that we had earlier started with that both the values have same average is false as the P value is very small. On checking the mean of the two")

print("\nThe mean for Naturally grown fruits =",round(Natural.Seed_Yield_Plant.mean(),2))
print("The mean for Hand pollination grown fruits =",round(Hand.Seed_Yield_Plant.mean(),2))

print("\nOn seeing the mean we can clearly see that the hand pollinated plants yield better seeds")


print("----------------------------------------------------------------------------------------------------------------------")


print("\nIn order to check for the fruit yield we'll be performing the 2 sample test on Seed Yield of plants column for both data\n")

print(stats.ttest_ind(a=Natural.Seedling_length,b=Hand.Seedling_length,equal_var=True))

print("\nBased on the P value we can conclude that the hypothesis that we had earlier started with that both the values have same average is false as the P value is small as compared to what it should be in case of 95% accuracy i.e 0.05 P value  On checking the mean of the two")

print("\nThe mean for Naturally grown fruits =",round(Natural.Seedling_length.mean(),2))
print("The mean for Hand pollination grown fruits =",round(Hand.Seedling_length.mean(),2))

print("\nOn seeing the mean we can clearly see that the hand pollinated plants have better seedling lengths")
```

```python
Dvdplayer=pd.read_csv("C:/Users/Vishal/Desktop/Python/7. Basic Stats - Case Study-1/dvdplayer.csv")

Dvdplayer.head()
```

```python
print("4. In order to check if all the age group rated it differently we'll consider the null hypothesis consideing that the average of all the age groups is same and check if this hypothesis holds true")

print("\n As we have 1 continuous and 1 catagorical with segments we'll be checking the hypothesis with ANOVA\n")

Age_25Less=Dvdplayer.dvdscore[Dvdplayer.agegroup=='Under 25']
Age_25_34=Dvdplayer.dvdscore[Dvdplayer.agegroup=='25-34']
Age_35_44=Dvdplayer.dvdscore[Dvdplayer.agegroup=='35-44']
Age_45_54=Dvdplayer.dvdscore[Dvdplayer.agegroup=='45-54']
Age_55_64=Dvdplayer.dvdscore[Dvdplayer.agegroup=='55-64']
Above_65=Dvdplayer.dvdscore[Dvdplayer.agegroup=='65 and over']


print(stats.f_oneway(Age_25Less,Age_25_34,Age_35_44,Age_45_54,Age_55_64,Above_65))

print("\nThe value of P is extremly small thus or initial assumption that all the averages are same doesn't hold true. On checking the means we can conclude")

print("\nFor age less than 25 the average is =",round(Age_25Less.mean(),2))
print("For age 25-34 the average is =",round(Age_25_34.mean(),2))
print("For age 35-44 the average is =",round(Age_35_44.mean(),2))
print("For age 45-54 the average is =",round(Age_45_54.mean(),2))
print("For age 55-64 the average is =",round(Age_55_64.mean(),2))
print("For age more than 65 the average is =",round(Above_65.mean(),2))

print("\nWe can clearly see that the averages of all the segments vary alot from each other")
```

```python
Sample_survey=pd.read_csv("C:/Users/Vishal/Desktop/Python/7. Basic Stats - Case Study-1/sample_survey.csv")

Sample_survey.head()
```

```python
print("5.a In order to check the relation between the  labour force status with marital status we'll be doing the Chi Sq test considering the fact that the observed frequency is equal to the expected frequency")

t=pd.crosstab(Sample_survey.wrkstat,Sample_survey.marital,margins=True)
print("\n",stats.chi2_contingency(observed=t))

print("\nAs we can see that the P value is extemely small our null hypothesis doent hold true thus we can conclude that the two columns are dependent oneach other as the variance of two is not independent")
```

```python
print("5.b In order to check if the educational qualification is controlling the marital status we'll be doing the Chi Sq test considering the fact that the observed frequency is equal to the expected frequency and the variance of the two is independent")

d=pd.crosstab(Sample_survey.marital,Sample_survey.degree,margins=True)
print("\n",stats.chi2_contingency(observed=d))

print("\nAs we can see that the P value is extemely small our null hypothesis doent hold true thus we can conclude that the two columns are dependent oneach other as the variance of two is not independent. Thus we can say that education qualification is somehow controlling the marital status")
```

```python
print("5.c.i In order to check if the happiness is driven by marital status we'll be doing the Chi Sq test considering the fact that the observed frequency is equal to the expected frequency and the variance of the two is independent")

e=pd.crosstab(Sample_survey.marital,Sample_survey.happy,margins=True)
print("\n",stats.chi2_contingency(observed=e))

print("\nAs we can see that the P value is extemely small our null hypothesis doent hold true thus we can conclude that the two columns are dependent oneach other as the variance of two is not independent. Thus we can say that the happiness is dependent on the marital status")
```
